##Usamos el método de Desviación estándar para eliminar outliers de las columnas con valores numéricos

```python
df2 = df1.copy()
```

```python
df2_numericos = df2.select_dtypes(include=['number'])
```

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=df2_numericos

Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
df3= df2_numericos[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df3
```

```python
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
df4 = df3.fillna(round(df3.mean(),1))
df4
```

##Aplicamos un análisis univariado a diferentes variables categóricas

Seleccionamos las variables categóricas

```python
freq_tbl(df5)
```

```python
var_cat = df5.select_dtypes(include=['object'])
```

```python
var_cat
```

```python
variables = var_cat.iloc[:, :20]

for variable in variables:
  print(variable)
print(len(variable))
```

```python
#Obtengo un análisis univariado de una variable categórica en específico
table1= freq_tbl(variables['184_store_devices'])
table1
```

```python
#Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
Filtro= table1[table1['frequency']>1]
Filtro
```

```python
Filtro_index= Filtro.set_index('184_store_devices')
Filtro_index
```

```python
Filtro.plot(kind = 'bar', width=1, figsize=(10,4))
plt.title('Herramientas de relación con los clientes')
plt.xlabel('Tipo de Herramienta')
plt.ylabel('Frecuencia')
```

```python
Filtro.plot("frequency", "cumulative_perc", kind="scatter")
```
